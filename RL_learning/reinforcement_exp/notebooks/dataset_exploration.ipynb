```python
augrid_csv_costs = os.path.join(DATA_DIR,'combined/augrid_costs.csv')
augrid_costs_df = pd.read_csv(augrid_csv_costs)
```

```python
augrid_costs_df.head()
```

```python
#average yearly cost per customer
augrid_costs_df['timestamp'] = pd.to_datetime(augrid_costs_df['timestamp'])
augrid_costs_df['year'] = augrid_costs_df['timestamp'].dt.year

avg_yearly_cost = (
    augrid_costs_df.groupby(['customer_id', 'year'])['import_cost_no_solar']
    .sum()
    .groupby('customer_id')
    .mean()
)
print(avg_yearly_cost.mean())

#avg solar used immediate
avg_yearly_cost = (
    augrid_costs_df.groupby(['customer_id', 'year'])['import_cost_solar_used_immediate']
    .sum()
    .groupby('customer_id')
    .mean()
)
print(avg_yearly_cost.mean())
```
